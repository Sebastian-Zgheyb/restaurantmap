```python
import requests
import json

url = "http://127.0.0.1:8000/get_data"  # Adjust if needed
payload = {
    "latitude": -33.88363285605243,
    "longitude": 151.21301531321515,
    "radius": 1500
}

response = requests.post(url, json=payload)

print("Status Code:", response.status_code)
print("Raw Response:", response.text)  # <-- Check this before calling .json()

try:
    response_data = response.json()  # This might fail
    print("Response JSON:", response_data)
    
    # Dump the response data to a JSON file
    with open('response_data.json', 'w') as json_file:
        json.dump(response_data, json_file, indent=4)
    print("Data has been written to 'response_data.json'.")

except requests.exceptions.JSONDecodeError:
    print("Error: Response is not valid JSON")
```
